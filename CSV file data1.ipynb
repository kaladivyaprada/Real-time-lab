```python
import os

file_path = "plan.txt"

if os.path.exists(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
        print(content)
else:
    # Create the file with some starter content
    with open(file_path, "w", encoding="utf-8") as file:
        file.write("This is a new plan file.\nAdd your content here.")
    print(f"File '{file_path}' was not found, so a new one has been created.")
```
